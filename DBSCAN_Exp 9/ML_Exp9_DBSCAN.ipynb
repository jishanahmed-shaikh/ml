```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1, 2], [2, 2], [2, 3],[8, 7], [8, 8], [25, 80]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
db = DBSCAN(eps=3, min_samples=2)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps=3, min_samples=3)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps=10, min_samples=2)
db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps=5, min_samples=2)
db.fit(X)
db.labels_
```

```python
from sklearn.datasets import make_circles
```

```python
# Create a concentric circle dataset
X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps=0.1, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
clusters
```

```python
# Plotting
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
#X[:, 0] and X[:, 1] represent the first and second features (coordinates) of the points in the 2D space.
#c=clusters colors the points according to their cluster labels, where points in the same cluster have the same color.
#cmap='viridis' specifies the colormap used for coloring the clusters.
plt.title("DBSCAN Clustering of Concentric Circles")
plt.show()
```
